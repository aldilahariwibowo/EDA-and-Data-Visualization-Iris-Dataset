# **Iris Dataset : Data Visualization**

```python
import matplotlib.pyplot as plt # data visualization with matplotlib
import seaborn as sns # data visualization with seaborn
```

**Bar Plot**

```python
iris_df['Species'].value_counts() # to count the number of each species
```

```python
iris_df['Species'].value_counts().plot.bar() # bar plot visualization using matplotlib
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=iris_df, x='Species') # bar plot visualization using seaborn
plt.tight_layout()
# sns counttplot
```

**Pie Chart**

```python
iris_df['Species'].value_counts().plot.pie(autopct= '%1.1f%%', labels=None, legend=True) # visualization of species data using a pie chart
plt.tight_layout()
```

**Line Plot**

```python
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(8,8)) # data visualization using a line chart based on categories

iris_df['SepalLengthCm'].plot.line(ax=ax[0][0])
ax[0][0].set_title('Sepal Length')

iris_df['SepalWidthCm'].plot.line(ax=ax[0][1])
ax[0][1].set_title('Sepal Width')

iris_df.PetalLengthCm.plot.line(ax=ax[1][0])
ax[1][0].set_title('Petal Length')

iris_df.PetalWidthCm.plot.line(ax=ax[1][1])
ax[1][1].set_title('Petal Width')
```

```python
iris_df.plot() # data visualization using line chart
plt.tight_layout()
```

**Histogram**

```python
iris_df.hist(figsize=(6,6), bins=10) # data visualization using histograms
plt.tight_layout()
```

Boxplot

```python
iris_df.boxplot() # data visualization using box plots
plt.tight_layout()
```

```python
iris_df.boxplot(by="Species", figsize=(8,8)) # data visualization using box plots group by species
plt.tight_layout()
```

**Scatter Plot**

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=iris_df, hue='Species') # data visualization using scatter plot with object of observation is the species
plt.tight_layout()
```

**Pair Plot**

```python
sns.pairplot(iris_df, hue='Species', markers='+') # data visualization using pair plot
plt.tight_layout()
```

**Heatmap**

```python
sns.heatmap(data=iris_df.corr()) # data visualization using heatmap
```

# **Conclusion**

The data set consists of 50 samples from each of three species of Iris (Iris Setosa, Iris virginica, and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters. The iris dataset has dimensions of 150 rows and 6 columns, but because the id column is not involved in the observation, the id column is removed from the list of columns, so that the dimensions of the iris dataset become 150 rows and 5 columns, namely sepal length, sepal width, petal length, petal width, species.

In the iris dataset there are 3 duplicate data, then the 3 duplicate data are removed and the data dimensions become 147 rows and 5 columns. Total of 147 species data consisting of 50 data of iris-versicolor, 49 data of iris-virginica, and 48 data of iris sentosa. Based on matrix correlation and supported by pair plot and heatmap visualization, the strongest correlation is petal width with petal length, so petal width greatly affects petal length
